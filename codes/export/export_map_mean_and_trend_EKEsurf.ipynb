```python
mesh = fesom_mesh(meshpath, get3d = get3d)
```

```python
VAR  = np.zeros((len(years),len(mesh.x2)))
U  = np.zeros((len(years),len(mesh.x2)))
V  = np.zeros((len(years),len(mesh.x2)))

for mo in range(3,9):
    print mo
    for ye in range(0,len(years)):
     print years[ye]
     ncfile  = resultpath+runid+'.'+str(years[ye])+'.oce.mean.nc'
     f1       = Dataset(ncfile, 'r')  
     ncfile  = resultpath+runid+'.'+str(years[ye])+'.oce.diag.nc'
     f2       = Dataset(ncfile, 'r')  
    
     if years[ye] < 2000:
        u     = f1.variables[var_id1][mo,0:mesh.n2d]
        v     = f1.variables[var_id2][mo,0:mesh.n2d]
        uu     = f2.variables[var_id3][mo,0:mesh.n2d]
        vv     = f2.variables[var_id4][mo,0:mesh.n2d]
        
        
     else:
        
        if mo ==3: 
            dayind = (range(45,60))
            month='APR'
        elif mo ==4: 
            dayind = (range(60,75))
            month='MAY'
        elif mo ==5: 
            dayind = (range(75,90))
            month='JUN'
        elif mo ==6: 
            dayind = (range(90,105))
            month='JUL'
        elif mo ==7:
            dayind = (range(105,120))
            month='AUG'
        elif mo ==8: 
            dayind = (range(120,135))
            month='SEP'
            
        u     = f1.variables[var_id1][dayind,0:mesh.n2d].mean(axis=0)
        v     = f1.variables[var_id2][dayind,0:mesh.n2d].mean(axis=0)
        uu     = f2.variables[var_id3][dayind,0:mesh.n2d].mean(axis=0)
        vv     = f2.variables[var_id4][dayind,0:mesh.n2d].mean(axis=0)
        
     ux, vy = pf.vec_rotate_r2g(50, 15, -90, mesh.x2, mesh.y2, u, v, flag=1)
     VAR[ye,:] = 10000*0.5*((uu+vv)-(ux**2+vy**2))  
    
    data2 = VAR.mean(axis=0)
    result_fit = np.ma.polyfit(years,VAR,1)
    trend = result_fit[0,:]
    
    if export_csv == True:
        np.savetxt(outputpath+'EKEsurf_'+month+'_'+str(first_year)+'_'+str(last_year)+'_timeseries.csv',VAR, delimiter=";")
        np.savetxt(outputpath+'EKEsurf_'+month+'_'+str(first_year)+'_'+str(last_year)+'_mean.csv', data2, delimiter=";")
        np.savetxt(outputpath+'EKEsurf_'+month+'_'+str(first_year)+'_'+str(last_year)+'_trend.csv', trend, delimiter=";")
        print 'exporting done ...'
```
