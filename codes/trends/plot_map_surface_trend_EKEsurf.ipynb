```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from pylab import *
import pyfesom as pf
#from load_mesh_data_new import *
import numpy as np
import datetime as dt
from PiecewiseNorm import PiecewiseNorm
from netCDF4 import Dataset
import cmocean as cmaps
#from matplotlib.colors import ListedColormap
```

```python
# Loading data from CSV Files
print 'loading...'
list=glob.glob(outputpath+'EKEsurf*_calctrend.csv')

if len(list) > 0:
    print 'trends already calculated'
    APR = genfromtxt(outputpath+'EKEsurf_APR_1990_2015_calctrend.csv', delimiter=';')
    print 'April...'
    MAY = genfromtxt(outputpath+'EKEsurf_MAY_1990_2015_calctrend.csv', delimiter=';')
    print 'May...'
    JUN = genfromtxt(outputpath+'EKEsurf_JUN_1990_2015_calctrend.csv', delimiter=';')
    print 'June...'
    JUL = genfromtxt(outputpath+'EKEsurf_JUL_1990_2015_calctrend.csv', delimiter=';')
    print 'July...'
    AUG = genfromtxt(outputpath+'EKEsurf_AUG_1990_2015_calctrend.csv', delimiter=';')
    print 'August...'
    SEP = genfromtxt(outputpath+'EKEsurf_SEP_1990_2015_calctrend.csv', delimiter=';')
    print 'September...'
else:
    print 'and calculating trend from yearly data'
    X = genfromtxt(outputpath+'EKEsurf_APR_1990_2015_trend.csv', delimiter=';')
    X=np.abs(X)
    result_fit = np.ma.polyfit(years,X,1)
    APR = result_fit[0,:]
    del X
    del result_fit
    np.savetxt(outputpath+'EKEsurf_APR_1990_2015_calctrend.csv', APR, delimiter=';')
    print 'April...'

    X = genfromtxt(outputpath+'EKEsurf_MAY_1990_2015_trend.csv', delimiter=';')
    X=np.abs(X)
    result_fit = np.ma.polyfit(years,X,1)
    MAY = result_fit[0,:]
    del X
    del result_fit
    np.savetxt(outputpath+'EKEsurf_MAY_1990_2015_calctrend.csv', MAY, delimiter=';')
    print 'May...'

    X = genfromtxt(outputpath+'EKEsurf_JUN_1990_2015_trend.csv', delimiter=';')
    X=np.abs(X)
    result_fit = np.ma.polyfit(years,X,1)
    JUN = result_fit[0,:]
    del X
    del result_fit
    np.savetxt(outputpath+'EKEsurf_JUN_1990_2015_calctrend.csv', JUN, delimiter=';')
    print 'June...'

    X = genfromtxt(outputpath+'EKEsurf_JUL_1990_2015_trend.csv', delimiter=';')
    X=np.abs(X)
    result_fit = np.ma.polyfit(years,X,1)
    JUL = result_fit[0,:]
    del X
    del result_fit
    np.savetxt(outputpath+'EKEsurf_JUL_1990_2015_calctrend.csv', JUL, delimiter=';')
    print 'July...'

    X = genfromtxt(outputpath+'EKEsurf_AUG_1990_2015_trend.csv', delimiter=';')
    X=np.abs(X)
    result_fit = np.ma.polyfit(years,X,1)
    AUG = result_fit[0,:]
    del X
    del result_fit
    np.savetxt(outputpath+'EKEsurf_AUG_1990_2015_calctrend.csv', AUG, delimiter=';')
    print 'August...'

    X = genfromtxt(outputpath+'EKEsurf_SEP_1990_2015_trend.csv', delimiter=';')
    X=np.abs(X)
    result_fit = np.ma.polyfit(years,X,1)
    SEP = result_fit[0,:]
    del X
    del result_fit
    np.savetxt(outputpath+'EKEsurf_SEP_1990_2015_calctrend.csv', SEP, delimiter=';')
    print 'September...'
```
